```python
# EXERCÍCIO 04

preco_combustivel = 5.00  # R$ 5,00 por litro
eficiencia_carro = 20  # km por litro

dinheiro_usuario = float(input("Quanto dinheiro você tem? R$ "))

litros_combustivel = dinheiro_usuario / preco_combustivel

quilometros_carro = litros_combustivel * eficiencia_carro

print(f"Você pode comprar {litros_combustivel:.2f} litros de combustível.")
print(f"Com essa quantidade de combustível, o carro consegue andar {quilometros_carro:.2f} quilômetros.")
```

```python
# EXERCÍCIO 05
v1 = 15
v2 = 10
v3 = 5
v4 = 0

print("a) (v1 = 10) AND (v2 = 10) =", (v1 == 10) and (v2 == 10))
print("b) (v1 = 15) AND (v2 = 10) =", (v1 == 15) and (v2 == 10))
print("c) (v4 = 1) OR (v3 = 4) =", (v4 == 1) or (v3 == 4))
print("d) (v1 > 10 AND v2 < 15) AND (v3 < 10 AND v4 = 0) =",
 (v1 > 10 and v2 < 15) and (v3 < 10 and v4 == 0))
print("e) (v1 < 10 AND v2 > 15) OR (v3 > 5 OR v4 = 0) =",
 (v1 < 10 and v2 > 15) or (v3 > 5 or v4 == 0))
```

```python
# Exercício 06
dias = int(input("Por favor, informe a quantidade de dias: "))

anos = dias // 360
meses = (dias % 360) // 30
dias_restantes = (dias % 360) % 30

print(f"A quantidade correspondente em anos, meses e dias é:\n{anos} ano(s)\n{meses} mês(es)\n{dias_restantes} dia(s)\n")
```
